```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv ('all_seasons.csv', index_col=0)
df.head()
```

```python
data = df.drop(['player_name', 'team_abbreviation', 'college', 'country', 'draft_year', 'draft_round', 'draft_number', 'season'], axis = 1)

data.head()
```

```python
features = ['age','player_height',	'player_weight', 'gp', 'pts', 'reb', 'ast', 'net_rating', 'oreb_pct', 'dreb_pct', 'usg_pct', 'ts_pct', 'ast_pct']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure (1, figsize = (25,10))
n=0
for x in features:
    n += 1
    plt.subplot(3, 5, n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.distplot(data[x], bins=15)
    plt.title('Distplot of {}'.format(x))

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X = data.values
scaler = StandardScaler()
scaleddata = scaler.fit_transform(X)

scaleddata
```

```python
if np.isnan(data).any() or np.isnan(scaleddata).any():
    print("Error: Your data contains NaN or missing values.")
else:
    # Rest of your code for plotting
    fig, ax = plt.subplots(1, 2, figsize=(15, 3))
    
    # Plotting for 'data'
    sns.histplot(data, ax=ax[0], kde=True, legend=False)
    ax[0].set_title("Data Awal")

    # Plotting for 'scaleddata'
    sns.histplot(scaleddata, ax=ax[1], kde=True, legend=False)
    ax[1].set_title("Standarisasi Data")

    plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaleddata)

data_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
print(data_pca)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data_pca['PC1'], data_pca['PC2'], alpha=0.5)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA Plot')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist


print("Pilih metode linkage:")
print("1. Single")
print("2. Complete")
print("3. Average")
print("4. Ward")
choice = int(input("Masukkan pilihan (1/2/3/4): "))


methods = ['single', 'complete', 'average', 'ward']
selected_method = methods[choice - 1]


distances = pdist(data, metric='euclidean')


Z = linkage(data, selected_method)

plt.figure(figsize=(12, 6))
plt.title(f'Dendrogram ({selected_method} linkage)')
dendrogram(Z, labels=data.index, orientation='top', truncate_mode='lastp', p=100)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import plotly.express as px
import plotly.graph_objects as go


n_clusters = 5

fig = go.Figure()

model = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
data['cluster_label'] = model.fit_predict(data)

for cluster_label in data['cluster_label'].unique():
    cluster_data = data[data['cluster_label'] == cluster_label]
    fig.add_trace(go.Scatter(x=cluster_data['BALANCE'], y=cluster_data['PURCHASES'], mode='markers', name=f'ward linkage'))

fig.update_layout(
    xaxis_title='Balance',
    yaxis_title='Purchase',
    title=f'Clusters of Customers Ward Linkage',
)

fig.show()
```
